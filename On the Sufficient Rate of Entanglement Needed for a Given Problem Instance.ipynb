# On the Sufficient Rate of Entanglement Needed for a Given Problem Instance

## Graph Instance

Consider the following $8$-nodes simle ring graph:

```python
import networkx as nx

num_nodes = 8
graph = nx.cycle_graph(num_nodes)
nx.draw(graph, with_labels=True)
```

## Trivial Solution

In the case of a ring graph, the optimal solution for the MaxCut problem is trivial:

```python
nodes_colors = ["green" if i % 2 == 0 else "lightblue" for i in range(num_nodes)]
nx.draw(graph, with_labels=True, node_color=nodes_colors)
```

## QEMC Analysis for the Graph Instance

### QEMC's Optimal Quantum State

Since the solution is trivial we can specify the optimal quantum state that the QEMC algorithm can generate to solve the problem-instance depicted above:

$$ |\psi_1\rangle = \frac{1}{2} \big( |000\rangle + |001\rangle + |010\rangle + |011\rangle \big) \tag{1} $$

\* The exact opposite state is also an optimal solution.

We can show that this state is separable:

$$|0\text{++}\rangle = |0\rangle \otimes \frac{1}{\sqrt{2}} \big( |0\rangle + |1\rangle \big) \otimes \frac{1}{\sqrt{2}}  \big( |0\rangle + |1\rangle \big) = \frac{1}{2} |0\rangle \otimes \big( |00\rangle + |01\rangle + |10\rangle + |11\rangle \big) = \frac{1}{2} \big( |000\rangle + |001\rangle + |010\rangle + |011\rangle \big) \tag{2} $$

$$ \big \downarrow $$

$$ |\psi_1\rangle = |0\text{++}\rangle = \frac{1}{2} \big( |000\rangle + |001\rangle + |010\rangle + |011\rangle \big) \tag{3} $$

Hence, on can conclude that no entanglement is needed to be expressed in the QEMC's Ansatz to achieve the optimal solution.

### QEMC Execution

Let's try to execute QEMC with the problem instance (in ideal noiseless conditions) with varying number of entangling layers:

```python
from qemc_implementation import QEMC

qemc_int = QEMC(graph)
ansatz = qemc_int.construct_ansatz(num_layers=7)
ansatz.draw("mpl", fold=-1)
```

```python
from qiskit_aer import StatevectorSimulator
from classical_functions import partition_draw

res = qemc_int.run(backend=StatevectorSimulator())
print(res)
partition_draw(graph, res.best_partition_bitstring, ax=None, pos=None)
```
