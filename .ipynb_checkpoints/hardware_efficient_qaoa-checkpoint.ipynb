```python
import networkx as nx

num_nodes = 16
degree = 9

graph = nx.random_graphs.random_regular_graph(degree, num_nodes)
pos = nx.circular_layout(graph)

nx.draw(graph, with_labels=True, pos=pos)
```

```python
from qiskit.circuit.library import RealAmplitudes, EfficientSU2

layers = 2

ansatz = EfficientSU2(num_qubits=num_nodes, reps=layers, insert_barriers=True).decompose()

ansatz.draw("mpl", fold=-1)
```

```python
from qaoa_implementation import MaxCutQAOA

qaoa_int = MaxCutQAOA(graph)

ansatz.measure_all()
res = qaoa_int.run(ansatz)
```
